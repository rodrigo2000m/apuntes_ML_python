## Límite:
El límite de $f(x)$ cuando $x$ tiene a $a$ es L.

$$
\lim_{x \rightarrow a} f(x) = L
$$

### Límites laterales

Límite lateral izquierdo

$$
\lim_{x \rightarrow a^-} f(x) = L
$$

Límite lateral derecho

$$
\lim_{x \rightarrow a^+} f(x) = L
$$

Sí los límites por derecha e izquierda son iguales entonces el limite de $f(x)$ cuando $x$ tiende a $a$ existe y vale $L$.


## Derivada

$$
f´(x) = \frac{dy}{dx}=\lim_{h \rightarrow 0} \frac{f(x+h)-f(x)}{h}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sin(x)

x = np.linspace(-10, 10, num=1000)
y = f(x)

def df(f):
    h=0.0000000001
    return (f(x+h)-f(x))/h

plt.plot(x,y, "red")
plt.plot(x, df(f), "blue")
plt.plot(x, np.cos(x), "yellow")
```

## La regla de la cadena

Para una función compuesta:

$$
y=f(g(x))
$$

La derivada se calcula como:

$$
\frac{dy}{dx}=\frac{dy}{dg} \cdot \frac{dg}{dx}
$$
